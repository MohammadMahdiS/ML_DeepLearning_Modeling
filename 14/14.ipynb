```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
x = np.arange(0, 1000)
```

```python
y = [ 2 * num + 3 for num in x]
```

```python
plt.plot(x, y, c = 'g')
```

```python
x_train, y_train = x[0:800], y[0:800]
x_test, y_test = x[800:], y[800:]
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
plt.plot(x_train, y_train, c = "r", label = 'Train')
plt.plot(x_test, y_test, c = "b", label = "Test")
plt.xlabel("X_data")
plt.ylabel("Y_data")
plt.legend()
plt.show()
```

# y = bx + a # b --> 2, a = 3 |

```python
model = tf.keras.models.Sequential(
    [
        tf.keras.layers.Dense(10),
        tf.keras.layers.Dense(5),
        tf.keras.layers.Dense(1),
    ]
)
```

```python
model.compile(optimizer = tf.keras.optimizers.SGD(lr=0.001) , loss = tf.keras.losses.mse() , )
```
